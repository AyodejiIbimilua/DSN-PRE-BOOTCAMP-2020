```python
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

style.use("ggplot")
```

```python
xs = np.array([2,4,6,7,8,9,12,13,5,7,16,18], dtype=np.float64)
ys = np.array([100,89,90,86,89,78,74,60,62,65,55,50], dtype=np.float64)
```

```python
def best_fit_slope_and_intercept(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
        ((mean(xs)**2) - mean(xs**2))
    )

    b = mean(ys) - m*(mean(xs))

    return m, b
```

```python
m, b = best_fit_slope_and_intercept(xs, ys)
```

```python
regression_line = [(m*x)+b for x in xs]
predict_x = 18
predict_y = (m * predict_x) + b
```

```python
plt.scatter(xs, ys)
plt.scatter(predict_x, predict_y, c="b")
plt.plot(xs, regression_line)
plt.show()
```
